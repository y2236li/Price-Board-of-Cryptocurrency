```python
from coinbase.wallet.client import Client
from coinbase.wallet.model import APIObject
from datetime import datetime

def Download ():
    client = Client('QKIpMN8aH0fIKz6A', 'VVEQmMdtLGCY15j58qUTaqszNpcI5n2')
    data = client._make_api_object(client._get('v2', 'prices', 'ETH-CAD', 'historic'), APIObject)
    data = pd.DataFrame(data)
    data['price'] = str(data.prices)
    tmp_string = data.prices
    price = []
    time = []
    date = []
    for i in tmp_string:
        i = str(i)
        i = i.split(" ")
        i[3] = i[3].split(",")[0]
        i[3] = i[3].replace('\"', '')
        tmp_date = i[6].split("T")[0]
        tmp_date = tmp_date.replace('\"', '')
        tmp_time = i[6].split("T")[1]
        tmp_time = tmp_time.split("Z")[0]
        tmp_time = datetime.strptime(tmp_time, '%H:%M:%S')
        jetlag = datetime.strptime('04:00:00', '%H:%M:%S')
        time.append(tmp_time-jetlag)
        date.append(tmp_date)
        price.append(i[3])
    data['price'] = price
    data['time'] = time
    data['date'] = date
    data.drop(['prices'], axis=1, inplace=True)
    return data
```
